```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = r'C:\Users\Mihir Patel\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_31753\API_SP.POP.TOTL_DS2_en_csv_v2_31753.csv'
population_data = pd.read_csv(file_path, skiprows=4)

# Drop columns with all NaN values and clean the data
population_data_clean = population_data.dropna(axis=1, how='all')

# Set 'Country Name' as the index
population_data_clean = population_data_clean.set_index('Country Name')

# Choose a year to visualize the population distribution (e.g., 2020)
year = '2020'
if year not in population_data_clean.columns:
    raise ValueError(f"Year {year} not found in the dataset")

# Get population data for the specified year, dropping NaN values
population_by_country = population_data_clean[year].dropna()

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(population_by_country, bins=30, edgecolor='black')
plt.title(f'Population Distribution in {year}')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.show()

# Create a bar chart for the top 10 most populated countries in the chosen year
top_10_countries = population_by_country.nlargest(10)

plt.figure(figsize=(10, 6))
top_10_countries.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title(f'Top 10 Most Populated Countries in {year}')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
